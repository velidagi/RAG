```python
# !pip install chromadb nltk sentence-transformers
import chromadb
from sentence_transformers import SentenceTransformer, CrossEncoder
import os
import pandas as pd
from nltk.tokenize import word_tokenize
import nltk
import numpy as np

# NLTK verilerini indir
nltk.download('punkt_tab')
nltk.download('punkt')

# Modelleri yükle
embedding_model = SentenceTransformer('all-MiniLM-L6-v2')
rerank_model = CrossEncoder('cross-encoder/ms-marco-MiniLM-L-6-v2')  # Rerank modeli

vektor_db_path = "/content/drive/MyDrive/Colab Notebooks/vector_db"

client = chromadb.PersistentClient(path=vektor_db_path)
collection = client.get_collection(name="notebook_documents")

def retrieve_relevant_documents(query, top_k=5):

    # Sorgu için embedding oluştur
    query_embedding = embedding_model.encode(query).tolist()

    # En benzer sonuçları sorgula
    results = collection.query(
        query_embeddings=[query_embedding],
        n_results=top_k
    )

    return results


def lexical_search(query, top_k=5):

    # Tüm belgeleri al
    all_documents = collection.get()
    documents = all_documents['documents']
    metadatas = all_documents['metadatas']
    ids = all_documents['ids']

    query_tokens = set(word_tokenize(query.lower()))


    matched_documents = []

    for i, doc in enumerate(documents):
        content_tokens = set(word_tokenize(doc.lower()))

        match_score = len(query_tokens.intersection(content_tokens))

        if match_score > 0:
            matched_documents.append({
                'document': doc,
                'metadata': metadatas[i],
                'id': ids[i],
                'score': match_score
            })

    matched_documents.sort(key=lambda x: x['score'], reverse=True)
    return matched_documents[:top_k]


def semantic_search_with_rerank(query, top_k=5, rerank_top_k=3):

    # Önce semantic search yap
    semantic_results = retrieve_relevant_documents(query, top_k=top_k)

    # Sonra rerank uygula
    reranked_results = apply_reranking(query, semantic_results, top_k=rerank_top_k)

    return reranked_results

def lexical_search_with_rerank(query, top_k=5, rerank_top_k=3):

    lexical_results = lexical_search(query, top_k=top_k)

    # Lexical sonuçları Chroma formatına dönüştür
    chroma_format_results = {
        'documents': [[item['document'] for item in lexical_results]],
        'metadatas': [[item['metadata'] for item in lexical_results]],
        'distances': [[0] * len(lexical_results)],  # Lexical search'te distance yok
        'ids': [[item['id'] for item in lexical_results]]
    }

    # Rerank
    reranked_results = apply_reranking(query, chroma_format_results, top_k=rerank_top_k)

    return reranked_results
```

```python
def hybrid_search(query, top_k=10, semantic_weight=0.7, lexical_weight=0.3):
    """
    Semantic ve lexical search sonuçlarını birleştirir
    """
    # Her iki arama türünü başlat
    semantic_results = retrieve_relevant_documents(query, top_k=top_k)
    lexical_results = lexical_search(query, top_k=top_k)

    # Benzersiz belgeleri ID'ye göre takip etmek için dictionary
    combined_docs = {}

    # Semantic sonuçları ekle (distance'ı score'a çevir)
    for i, (doc, metadata, doc_id, distance) in enumerate(zip(
        semantic_results['documents'][0],
        semantic_results['metadatas'][0],
        semantic_results['ids'][0],
        semantic_results['distances'][0]
    )):

        semantic_score = 1 / (1 + distance) if distance > 0 else 1.0
        combined_score = semantic_score * semantic_weight

        combined_docs[doc_id] = {
            'document': doc,
            'metadata': metadata,
            'id': doc_id,
            'semantic_score': semantic_score,
            'lexical_score': 0,
            'combined_score': combined_score
        }

    # Lexical sonuçları ekle veya güncelle
    for result in lexical_results:
        doc_id = result['id']
        lexical_score = result['score'] / 10  # Normalize et (örnek normalizasyon)

        if doc_id in combined_docs:
            # Zaten varsa, lexical score'u ekle
            combined_docs[doc_id]['lexical_score'] = lexical_score
            combined_docs[doc_id]['combined_score'] += lexical_score * lexical_weight
        else:
            # Yeni belge ekle
            combined_docs[doc_id] = {
                'document': result['document'],
                'metadata': result['metadata'],
                'id': doc_id,
                'semantic_score': 0,
                'lexical_score': lexical_score,
                'combined_score': lexical_score * lexical_weight
            }

    # Sıralanmış liste oluştur
    sorted_results = sorted(combined_docs.values(), key=lambda x: x['combined_score'], reverse=True)

    return sorted_results[:top_k]
```

```python
def hybrid_search_with_rerank(query, top_k=10, rerank_top_k=5, semantic_weight=0.7, lexical_weight=0.3):
    # Hybrid search yap
    hybrid_results = hybrid_search(query, top_k=top_k,
                                 semantic_weight=semantic_weight,
                                 lexical_weight=lexical_weight)

    # Rerank için Chroma formatına dönüştür
    chroma_format_results = {
        'documents': [[item['document'] for item in hybrid_results]],
        'metadatas': [[item['metadata'] for item in hybrid_results]],
        'distances': [[1 - item['combined_score'] for item in hybrid_results]],  # uzaklığı 1'e yakın olanın skoru en kötü.
        'ids': [[item['id'] for item in hybrid_results]]
    }

    # Rerank uygula
    reranked_results = apply_reranking(query, chroma_format_results, top_k=rerank_top_k)

    return {
        'hybrid_results': hybrid_results,
        'reranked_results': reranked_results
    }
```

```python
def apply_reranking(query, results, top_k):
    """
    Cross-encoder ile reranking uygula (geliştirilmiş)
    """
    documents = results['documents'][0]
    metadatas = results['metadatas'][0]
    distances = results['distances'][0] if 'distances' in results else [0] * len(documents)
    ids = results['ids'][0]
    original_scores = results.get('scores', [1.0] * len(documents))

    pairs = [(query, doc) for doc in documents]

    # Cross-encoder
    rerank_scores = rerank_model.predict(pairs)

    # Orijinal skorlarla birleştir (isteğe bağlı)
    final_scores = []
    for i, (rerank_score, original_score) in enumerate(zip(rerank_scores, original_scores)):
        # Rerank skorunu ağırlıklı olarak kullan, orijinal skoru da dikkate al
        combined_score = 0.8 * rerank_score + 0.2 * original_score
        final_scores.append(combined_score)

    ranked_indices = np.argsort(final_scores)[::-1]  # yüksekten -> düşüğe

    reranked_results = {
        'documents': [[documents[i] for i in ranked_indices[:top_k]]],
        'metadatas': [[metadatas[i] for i in ranked_indices[:top_k]]],
        'distances': [[distances[i] for i in ranked_indices[:top_k]]],
        'ids': [[ids[i] for i in ranked_indices[:top_k]]],
        'rerank_scores': [final_scores[i] for i in ranked_indices[:top_k]],
        'original_scores': [original_scores[i] for i in ranked_indices[:top_k]]
    }

    return reranked_results
```

```python
def test_hybrid_search(query):
    print("=== HYBRID SEARCH (Semantic + Lexical) ===")
    hybrid_results = hybrid_search(query, top_k=8)
    for i, result in enumerate(hybrid_results[:5]):
        print(f"{i+1}. [Combined: {result['combined_score']:.3f}, "
              f"Semantic: {result['semantic_score']:.3f}, "
              f"Lexical: {result['lexical_score']:.3f}] "
              f"{result['document']}... (Source: {result['metadata']['source']})")

    print("\n=== HYBRID SEARCH + RERANK ===")
    hybrid_reranked = hybrid_search_with_rerank(query, top_k=8, rerank_top_k=5)
    reranked_results = []

    for i, (doc, metadata, score) in enumerate(zip(
        hybrid_reranked['reranked_results']['documents'][0],
        hybrid_reranked['reranked_results']['metadatas'][0],
        hybrid_reranked['reranked_results']['rerank_scores']
    )):
        print(f"{i+1}. [Rerank Score: {score:.4f}] {doc}... (Source: {metadata['source']})")

        reranked_results.append({
            'rank': i+1,
            'document': doc,
            'metadata': metadata,
            'rerank_score': score
        })

    return reranked_results
# Test
query = "Where is capital Zimbabwe"
final_result = test_hybrid_search(query)

print("fonskiyon sonrası:",final_result)
```

```python
# !pip install google-generativeai
import google.generativeai as genai
from typing import List, Dict
from google.colab import userdata

GEMINI_API_KEY = userdata.get('GOOGLE_API_KEY')

genai.configure(api_key=GEMINI_API_KEY)

model = genai.GenerativeModel('gemini-2.0-flash')


def ask_llm(query, context_documents):
    context_text = "\n\n".join([f"Document {i+1}: {doc}" for i, doc in enumerate(context_documents)])

    prompt = f"""You are an assistant that answers questions about countries using only the provided context documents. These documents contain information about 195 countries from Wikipedia.

INSTRUCTIONS:
- Answer the question using ONLY the information provided in the context documents below
- If the answer cannot be found in the provided documents, respond with "I cannot find this information in the provided context"
- Do not use any external knowledge beyond what is given in the context
- Be specific and cite which document(s) you're referencing when possible
- If information is partial or unclear in the context, acknowledge this limitation

CONTEXT DOCUMENTS:
{context_text}

QUESTION: {query}

ANSWER:"""

    response = model.generate_content(prompt)
    return response.text

def rag_pipeline(query):
    print("RAG Pipeline Çalışıyor...")

    # Hybrid search(semantic+lexical) + rerank
    results = test_hybrid_search(query)

    # sorgulama sonucu

    context_documents = [result['document'] for result in results]

    print("\n--- LLM SORGULANIYOR ---")
    answer = ask_llm(query, context_documents)

    print("\n--- LLM CEVABI ---")
    print(answer)

    return answer


query = "Which country is the largest in terms of surface area in the world?"
answer = rag_pipeline(query)
```
